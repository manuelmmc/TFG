```python
import pandas as pd
import scipy.stats as stats

# Cargar los datos desde un archivo CSV
data = pd.read_csv('C:\\Users\\manue\\combined_model_performance.csv')

# Mostrar los datos originales para verificar
print("Datos originales:")
print(data.head())

# Convertir datos a numéricos, coerciendo errores a NaN en las columnas seleccionadas
data_numeric = data.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')  # Ignora la primera columna

# Mostrar datos después de la conversión
print("Datos después de la conversión a numéricos:")
print(data_numeric.head())

# Realizar la prueba de Shapiro-Wilk para cada columna si hay datos suficientes
normality_results = {}
for column in data_numeric.columns:
    if len(data_numeric[column].dropna()) > 3:  # Verificar si hay suficientes datos no-NaN para realizar la prueba
        stat, p_value = stats.shapiro(data_numeric[column].dropna())
        normality_results[column] = {'statistic': stat, 'p_value': p_value}
    else:
        normality_results[column] = {'statistic': None, 'p_value': None}

# Mostrar los resultados de la normalidad
print("Normalidad de los datos:")
for model, result in normality_results.items():
    if result['statistic'] is not None:
        print(f"{model}: Estadístico={result['statistic']}, P-Valor={result['p_value']}")
    else:
        print(f"{model}: Datos insuficientes para la prueba")

# Prueba de Levene para homogeneidad de varianzas si hay datos suficientes
if all(data_numeric[column].dropna().size > 0 for column in data_numeric.columns):
    stat, p_value = stats.levene(*[data_numeric[column].dropna() for column in data_numeric.columns])
    print(f"Levene Statistic: {stat}, P-Value: {p_value}")
else:
    print("No es posible realizar la prueba de Levene debido a datos insuficientes")
```

```python
import pandas as pd
import scipy.stats as stats

# Cargar los datos desde un archivo CSV
data = pd.read_csv('C:\\Users\\manue\\influences_combined.csv')

# Mostrar los datos originales para verificar
print("Datos originales:")
print(data.head())

# Convertir datos a numéricos, coerciendo errores a NaN en las columnas seleccionadas
data_numeric = data.iloc[:, 1:].apply(pd.to_numeric, errors='coerce')  # Ignora la primera columna

# Mostrar datos después de la conversión
print("Datos después de la conversión a numéricos:")
print(data_numeric.head())

# Realizar la prueba de Shapiro-Wilk para cada columna si hay datos suficientes
normality_results = {}
for column in data_numeric.columns:
    if len(data_numeric[column].dropna()) > 3:  # Verificar si hay suficientes datos no-NaN para realizar la prueba
        stat, p_value = stats.shapiro(data_numeric[column].dropna())
        normality_results[column] = {'statistic': stat, 'p_value': p_value}
    else:
        normality_results[column] = {'statistic': None, 'p_value': None}

# Mostrar los resultados de la normalidad
print("Normalidad de los datos:")
for model, result in normality_results.items():
    if result['statistic'] is not None:
        print(f"{model}: Estadístico={result['statistic']}, P-Valor={result['p_value']}")
    else:
        print(f"{model}: Datos insuficientes para la prueba")

# Prueba de Levene para homogeneidad de varianzas si hay datos suficientes
if all(data_numeric[column].dropna().size > 0 for column in data_numeric.columns):
    stat, p_value = stats.levene(*[data_numeric[column].dropna() for column in data_numeric.columns])
    print(f"Levene Statistic: {stat}, P-Value: {p_value}")
else:
    print("No es posible realizar la prueba de Levene debido a datos insuficientes")
```
